## **The (Hilbert) de Rham Complex & Poincaré operators**

We all know and love the de Rham complex:

$$H^1_0 \xrightarrow{\text{grad}} H_0(\text{curl}) \xrightarrow{\text{curl}} H_0(\text{div}) \xrightarrow{\text{div}} L^2_0.$$

It's obvious this is a *complex*:
- $\text{curl}\circ\text{grad} = 0$;
- $\text{div}\circ\text{curl} = 0$.

It's also pretty clear these maps are *well-defined*:
- $\text{grad}\,H^1_0 \subseteq H_0(\text{curl})$;
- $\text{curl}\,H_0(\text{curl}) \subseteq H_0(\text{div})$;
- $\text{div}\,H_0(\text{div}) \subseteq L^2_0$.

What's not obvious is that it's *exact**:
- $\text{grad}\,H^1_0 = {\cal N}(\text{curl}) \subseteq H_0(\text{curl})$;
- $\text{curl}\,H_0(\text{curl}) = {\cal N}(\text{div}) \subseteq H_0(\text{div})$;
- $\text{div}\,H_0(\text{div}) = L^2_0$.

Those definitions are a little notation heavy;
we can state them equivalently as:
- For ${\bf u} \in H_0(\text{curl})$, if $\text{curl}\,{\bf u} = {\bf 0}$ then there exists some $h_\text{grad}[{\bf u}] \in H^1_0$ such that $\text{grad}\,h_\text{grad}[{\bf u}] = {\bf u}$;
- For ${\bf u} \in H_0(\text{div})$, if $\text{div}\,{\bf u} = 0$ then there exists some $h_\text{curl}[{\bf u}] \in H_0(\text{curl})$ such that $\text{curl}\,h_\text{curl}[{\bf u}] = {\bf u}$;
- For $u \in L^2_0$, there exists some $h_\text{div}[u] \in H_0(\text{div})$ such that $\text{div}\,h_\text{div}[u] = u$;

Proving exactness requires constructing *Poincaré operators*:

>  **Poincaré operator** */'pwæ̃kɑːreɪ 'ɒpəɹeɪtə/ n.* (Right) inverses, that are well-defined on differential operator kernels.

These are precisely the $h_\bullet$'s I just implicitly mentioned above.
They're operators which undo differential operators: as close as you can get to inverse operators (given differential operators are in general neither surjective nor injective).

> **On topologically trivial domains.*

### Non-uniqueness of Poincaré operators

If such Poincaré operators exist*, they're *far* from unique:
We may add any element from the range of the previous operator without affecting the inverse property.

Thus to show a complex is exact, we need only construct one such Poincaré operator for each stage.

> **Spoiler: They do.*

### The minimisation approach

A systematic way of constructing Poincaré operators is through variational principles.
For example, to construct a Poincaré operator acting as an "inverse" on $\text{div}$, we may consider the problem:

> Minimise $\|{\bf h}\|_{L^2}$ subject to $\text{div}\,{\bf h} = u$.

To construct a Poincaré operator acting as an "inverse" on $\text{curl}$, we may consider the problem:

> Minimise $\|{\bf h}\|_{L^2}$ subject to $\text{curl}\,{\bf h} = {\bf u}$.

Etc. etc...

These minimisation problems may be written weakly through Lagrange multipliers, e.g. for the Poincaré operator inverting $\text{div}$:

> Find $({\bf h}, p) \in H_0(\text{div}) \times L^2_0$ such that
>
> $$({\bf h}, v) + (p, \text{div}\,v) = 0, \qquad (\text{div}\,{\bf h}, q) = (u, q),$$
>
> for all $(v, q) \in H_0(\text{div}) \times L^2_0$.

Any variational system like this can be converted to a PDE in strong form*, in this case:

> Find ${\bf h}$, $p$ such that
>
> $${\bf h} - \text{grad}\,p = 0, \qquad \text{div}\,{\bf h} = u,$$
>
> subject to the BC
>
> $${\bf h}\cdot{\bf n} = 0.$$

For each stage in a complex therefore, exactness is then equivalent to the the well-posedness of a certain PDE problem.
No surprise then, issues like boundary regularity pop in (as well as the familiar issues of topology/homology).

Note:
The (Hilbert) de Rham complex is exact on topologically trivial domains with Lipschitz boundaries.
Just trust me on this one.

These PDE problems can look very fimiliar;
in the case above, this looks like the Darcy flow problem!
We can think of these problems, in some sense, as the canonical linear problems associated with a complex.

> **We're just used to doing this in reverse: strong $\rightarrow$ weak, not weak $\rightarrow$ strong.*

## **(Hilbert) de Rham $\rightarrow$ (Hilbert) Stokes**

The Stokes/Navier–Stokes equations concern incompressible/divergence-free flows.
This naturally suggests formulating the problem within the kernel of $\text{div}$ in the above de Rham complex.

However, consider the canonical weak formulation of Stokes*:

> Find $({\bf u}, p)$ such that
>
> $$(\text{grad}\,{\bf u}, \text{grad}\,{\bf v}) + (p, \text{div}\,{\bf v}) = ({\bf f}, {\bf v}), \qquad (\text{div}\,{\bf u}, q) = 0,$$
>
> for all $({\bf v}, q)$.

The space $H_0(\text{div})$ isn't sufficiently regular for this weak formulation to be valid: $H_0(\text{div}) \not\subseteq H^1_0$.

> **Technically, this $\text{grad}$ should be the symmetric gradient $\text{sym}\,\text{grad}$. For our purposes today, the two are identical. You only need to start worrying when you consider slip BCs, non-conforming complexes, or non-flat manifolds.*

### Upgrading to $H^1_0$

A natural question arises:

> Can we construct a complex that *looks like the de Rham complex*, but where that penultimate space is $H^1_0$ instead?

We're theorising something that looks like:

$$? \xrightarrow{\text{grad}} \;? \xrightarrow{\text{curl}} H^1_0 \xrightarrow{\text{div}} L^2_0.$$

Well, as established above, we can theoretically choose whatever spaces we want along our complex, provided the associated linear problems remain well-posed.
Therefore, the feasibility of upgrading to $H^1$ reduces to the question:

> Is the associated linear problem at $\text{div}$ well-posed?

What does that problem look like?
In weak form:

> Find $({\bf u}, p) \in H^1_0 \times L^2_0$ such that
>
> $$(\text{grad}\,{\bf u}, \text{grad}\,{\bf v}) + (p, \text{div}\,{\bf v}) = ({\bf f}, {\bf v}), \qquad (\text{div}\,{\bf u}, q) = 0,$$
>
> for all $({\bf v}, q) \in H^1_0 \times L^2_0$.

And in strong form:

> Find ${\bf u}$, $p$ such that
>
> $$- \Delta{\bf u} - \text{grad}\,p = {\bf f}, \qquad \text{div}\,{\bf u} = 0,$$
>
> subject to the BC
>
> $${\bf u} = {\bf 0}.$$

*It's Stokes!*
We're safe therefore to upgrade this regularity whenever Stokes is well-posed, which is still Lipschitz domains with these BCs.

### Regularity propagation back through the complex

Everything else in the complex works fine, with the small caveat that $H_0(\text{curl})$ gets subbed out for $H_0(\text{grad}\,\text{curl})$,

$$H_0(\text{grad}\,\text{curl}) = \{{\bf u} \in H_0(\text{curl})\;\text{s.t.}\;\text{curl}\,{\bf u} \in H^1_0\}.$$

This is just so that $\text{curl}\,H_0(\text{grad}\,\text{curl}) \subseteq H^1_0$;
no fancier reason.
We then get the complex:

$$H^1_0 \xrightarrow{\text{grad}} H_0(\text{grad}\,\text{curl}) \xrightarrow{\text{curl}} H^1_0 \xrightarrow{\text{div}} L^2_0.$$

*Wild.*
This is called the Stokes complex.

> **Stokes complex** */stəʊks 'kɒmplɛks/ n.* Essentially just the de Rham complex, but with some enhanced regularity such that we end with $H^1_0 \rightarrow L^2_0$.

Crucially, it's the natural home for Stokes/Navier–Stokes.
The cohomology groups are just the same as those of the original de Rham complex too.

In 2D, things are a touch simpler:

$$H^2_0 \xrightarrow{\text{curl}} H^1_0 \xrightarrow{\text{div}} L^2_0.$$

*So why have I shown you all this?*
Well, when constructing new elemenets (and in general in life) it's important for us to think in terms of complexes when possible.
So the hunt for conforming elements for Stokes/Navier–Stokes starts at discretising the Stokes complex.

## **Discrete complexes of polynomials**

Say we're considering some potential discrete complex,

$$\mathbb{A} \xrightarrow{\text{grad}} \mathbb{B} \xrightarrow{\text{curl}} \mathbb{C} \xrightarrow{\text{div}} \mathbb{D}.$$

Showing we've constructed a discrete complex follows pretty much the same story.
Again, it's obvious this is a *complex*:
- $\text{curl}\circ\text{grad} = 0$;
- $\text{div}\circ\text{curl} = 0$.

We then need to show these maps are *well-defined*:
- $\text{grad}\,\mathbb{A} \subseteq \mathbb{B}$;
- $\text{curl}\,\mathbb{B} \subseteq \mathbb{C}$;
- $\text{div}\,\mathbb{C} \subseteq \mathbb{D}$.

We then need to show *exactness* by constructing Poincaré operators:
- For ${\bf u} \in \mathbb{B}$, if $\text{curl}\,{\bf u} = {\bf 0}$ then there exists some $h_\text{grad}[{\bf u}] \in \mathbb{A}$ such that $\text{grad}\,h_\text{grad}[{\bf u}] = {\bf u}$;
- For ${\bf u} \in \mathbb{C}$, if $\text{div}\,{\bf u} = 0$ then there exists some $h_\text{curl}[{\bf u}] \in \mathbb{B}$ such that $\text{curl}\,h_\text{curl}[{\bf u}] = {\bf u}$;
- For $u \in \mathbb{D}$, there exists some $h_\text{div}[u] \in \mathbb{C}$ such that $\text{div}\,h_\text{div}[u] = u$;

### Poincaré operators for de Rham complexes

Over a general simplicial mesh, let's fix ideas by considering the FE complex $\mathbb{P}_{p,\bullet}$,

$$\mathbb{CG}_{p+3} \xrightarrow{\text{grad}} \mathbb{NED}^{\text{curl}, \text{B}}_{p+2} \xrightarrow{\text{curl}} \mathbb{NED}^{\text{div}, \text{B}}_{p+1} \xrightarrow{\text{div}} \mathbb{DG}_p.$$

A neat way to form these spaces is by intersecting the continuous de Rham complex,

$$H^1_0 \xrightarrow{\text{grad}} H_0(\text{curl}) \xrightarrow{\text{curl}} H_0(\text{div}) \xrightarrow{\text{div}} L^2_0,$$

with polynomials of degree $p+3, \dots, p$ over our mesh.
One of the most precious tools in FEEC is that of *commuting projections* $\pi_\bullet$:

> **Commuting projections** */kəˈmjuːtɪŋ pɹəˈd͡ʒɛkʃən/ n.* Projections from a continuous complex to a discrete complex that commute.

\begin{align*}
    &H^1_0 &&\xrightarrow{\text{grad}} &&H_0(\text{curl}) &&\xrightarrow{\text{curl}} &&H_0(\text{div}) &&\xrightarrow{\text{div}} &&L^2_0  \\
    \pi&\downarrow\uparrow\iota &&& \pi&\downarrow\uparrow\iota &&& \pi&\downarrow\uparrow\iota &&& \pi&\downarrow\uparrow\iota  \\
    &\mathbb{CG}_{p+3} &&\xrightarrow{\text{grad}} &&\mathbb{NED}^{\text{curl}, \text{B}}_{p+2} &&\xrightarrow{\text{curl}} &&\mathbb{NED}^{\text{div}, \text{B}}_{p+1} &&\xrightarrow{\text{div}} &&\mathbb{DG}_p
\end{align*}

*Commuting projections give us discrete Poincaré operators for free!*
\begin{align*}
    \pi\circ h_\text{grad}\circ\iota, &&
    \pi\circ h_\text{curl}\circ\iota, &&
    \pi\circ h_\text{div}\circ\iota.
\end{align*}

The proof's simple enough.
Regardless, the upshot is as follows:
> *If we have commuting projections, we have Poincaré operators.*
> *If we have Poincaré operators, we have exactness.*

These projections certainly exist for $\mathbb{P}_{p,\bullet}$, so we know this forms a discrete complex.

### Poincaré operators for Stokes complexes

Now, this story is a little more dire...

Over a general simplicial mesh, let's hypothesise a discrete Stokes complex

$$\mathbb{CG}_{p+3} \xrightarrow{\text{grad}} \;?_{p+2} \xrightarrow{\text{curl}} [\mathbb{CG}_{p+1}]^3 \xrightarrow{\text{div}} \mathbb{DG}_p,$$

formed by intersecting the discrete Stokes complex,

$$H^1_0 \xrightarrow{\text{grad}} H_0(\text{grad}\,\text{curl}) \xrightarrow{\text{curl}} H^1_0 \xrightarrow{\text{div}} L^2_0,$$

with polynomials of degree $p+3, \dots, p$.

It's simple enough to see that *exactness breaks at $\text{div}$* by confirming that a Poincaré operator *$h_\text{div}$ cannot exist*.
The argument in 2D is just the same as in 3D, so let's stick to 2D.
Let our mesh just be a single triangle.

[triangle]

Consider ${\bf u} \in [\mathbb{CG}_{p+1}]^3 \cap H^1_0$, i.e. degree-$(p+1)$ polynomial ${\bf u}$ such that ${\bf u} = {\bf 0}$ on the boundary.

[triangle with DoFs]

In the corner, ${\bf u} = {\bf 0}$ along 2 (non-tangential) lines;
the only way this is possible in polynomials is if $\text{grad}\,{\bf u} = 0$ at that corner.

[corner highlighted]

This then implies $\text{div} (= \text{tr}[\text{grad}\,{\bf u}]) = 0$ at the corner.
So, if I have some $p \in \mathbb{DG}_p$ such that $p \ne 0$ at the corner, there is no way I can find $h_\text{div}[p] \in [\mathbb{CG}_{p+1}]^3 \cap H^1_0$ such that $\text{div}\,h_\text{div}[p] = p$.

The upshot of all this is as follows:
> *There is no discrete Stokes complex of polynomials on arbitrary simplicial meshes.*

## **Finite elements for the Stokes complex**

We're left with a choice:
- Either consider non-polynomials...
- ...or consider only certain special meshes.

To consider our options, let's restrict our attention to 2D.
(3D is a *whole* different story!)

### Alfeld splits: $\mathbb{HCT}_3$ $\rightarrow$ $[\mathbb{CG}^\text{micro}_2]^2$ $\rightarrow$ $\mathbb{DG}_1^\text{micro}$

Alfeld splits, or barycentric refinement, is a simple idea.
You just take your mesh...

Screenshot_20251112_232604_OneNote.jpg

...put a point in the barycentre (i.e. centre of mass) of each triangle...

Screenshot_20251112_232713_OneNote.jpg

...then split each cell into 3 about that point...

Screenshot_20251112_233814_OneNote.jpg

We then just use that as our new mesh.
The original triangles (composed now of 3 smaller triangles) are referred to as *macro-elements*;
the smaller triangles are similarly referred to as *micro-elements*.

Something magical happens though when we do this:

> *The Stokes complex has Poincaré operators for polynomials on Alfeld-split meshes. ([Fu, Guzmán & Neilan, 2020](https://arxiv.org/abs/1807.05883))*

Intersecting the discrete Stokes complex with polynomials of degree $3, 2, 1$ gives you the complex,

$$\mathbb{HCT}_3 \xrightarrow{\text{curl}} [\mathbb{CG}^\text{micro}_2]^2 \xrightarrow{\text{div}} \mathbb{DG}_1^\text{micro}.$$

When thinking about the DoFs for these spaces, one should interpret them as being defined on the *macro-elements*.
where:
- $\mathbb{HCT}_3 =$ the [Hsieh–Clough–Tocher element](https://defelement.org/elements/hsieh-clough-tocher.html): *this does not have a local basis on the micro-elements, only on the macro-elements, and therefore requires the Alfeld split to be defined*;
- $\mathbb{CG}^\text{micro}_2 =$ just [CG](https://defelement.org/elements/lagrange.html) on each micro-element;
- $\mathbb{DG}^\text{micro}_1 =$ just [DG](https://defelement.org/elements/discontinuous-lagrange.html) on each micro-element.

Screenshot_20251112_234917_OneNote.jpg

### Scrapping DoFs: $\mathbb{BELL}_2$ $\rightarrow$ $\mathbb{GN}^\text{A}_1$ $\rightarrow$ $\mathbb{DG}_0$

What if we want to decrease the order, i.e. decrease the number of DoFs on each element?
It's nice that we made the previous space using friendly polynomials, but it would be nice if we could remove some redundant basis functions and make these spaces a little smaller, without breaking the complex property.

A general rule of thumb for when you can do this is as follows:
> *When looking at a FE complex, we can remove a DoF on a given vertex/edge/cell, if we do the same for either the preceeding or following element in the complex.*

This isn't always true, and we do have to be careful we remove the right DoFs, but the general idea will do for our purposes.

1. Those gradient evaluations on the edges in $\mathbb{HCT}_3$ aren't doing anything;
let's scrap them.
To keep this a complex, we also need to scrap the tangential component evaluations on the edge of $[\mathbb{CG}^\text{micro}_2]^2$.

Screenshot_20251112_235021_OneNote.jpg

2. In fact, none of the 8 evaluations on the interior of $[\mathbb{CG}^\text{micro}_2]^2$ are necessary
let's scrap them all.
Similarly, we'll need to scrap the 8 of the (interior) DoFs in $\mathbb{DG}^\text{micro}_1$.

Screenshot_20251112_235109_OneNote.jpg

This leaves us with the following complex:

$$\mathbb{BELL}_2 \xrightarrow{\text{curl}} \mathbb{GN}^\text{A}_1 \xrightarrow{\text{div}} \mathbb{DG}_0$$

- $\mathbb{BELL}_2 =$ the [Bell element](https://defelement.org/elements/bell.html);
- $\mathbb{GN}^\text{A}_1 =$ the [type-A Guzmán–Neilan element](https://defelement.org/elements/guzman-neilan.html);
- $\mathbb{DG}_0 =$ just [DG](https://defelement.org/elements/discontinuous-lagrange.html).

Unless I'm mistaken* this is as few DoFs as you can get.

So why type-A**?
Because there's a type-B too of course!

> ***This is not common terminology. I'm just using here for disctinction.*

### Rational enhancement: $?_2$ $\rightarrow$ $\mathbb{GN}^\text{B}_1$ $\rightarrow$ $\mathbb{DG}_0$

Another way to view/derive the above complex is as follows:
1. Start with polynomials of degree $2$, $1$ and $0$;
2. Observe that that's not going to work, because we can't get a non-zero gradient in the corner (as above);
3. Realise functions on Alfeld splits can have non-zero gradients in the corner;
4. Enrich the first 2 spaces with a few new shape functions defined on Alfeld splits.

Alternatively, it's quite simple to find a rational function that's $0$ on the boundary of a cell, but has a non-zero gradient in the corner.
Enriching with these instead gives the following:

$$?_2 \xrightarrow{\text{curl}} \mathbb{GN}^\text{B}_1 \xrightarrow{\text{div}} \mathbb{DG}_0$$

- $\mathbb{?}_2 =$ *I don't know...? I don't know anyhwere this is considered outside the Guzmán–Neilan paper from 2014...?*;
- $\mathbb{GN}^\text{B}_1 =$ the [type-B Guzmán–Neilan element](https://www.jstor.org/stable/pdf/24488200);
- $\mathbb{DG}_0 =$ just [DG](https://defelement.org/elements/discontinuous-lagrange.html).

The DoFs are all in exactly the same place, it's just a slightly different basis.
