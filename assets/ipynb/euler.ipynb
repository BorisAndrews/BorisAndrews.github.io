# **Euler vs. the inviscid limit of Navier–Stokes**

> ***The Euler equations cannot be conflated with the Navier–Stokes equations in the limit of vanishing viscosity ($\text{Re} \to \infty$ or $\nu \to 0_+$).***

They may seem identical when we just look at either as a set of symbols in a strong-form PDE, but analytically they represent two very distinct regimes.
It is crucial to distinguish between them:

A. **The inviscid limit of Navier–Stokes:**
The accumulation point(s) of sequence of solutions $(\mathbf{u}_\nu)$ to Navier–Stokes $\dot{\mathbf{u}} = - \mathbf{u}\cdot\nabla\mathbf{u} - \nabla p + \nu\Delta\mathbf{u}$ as $\nu \to 0$.

B. **The Euler equations:**
Solutions $\mathbf{u}$ *(if they exist...)* to the ideal system $\dot{\mathbf{u}} = - \mathbf{u}\cdot\nabla\mathbf{u} - \nabla p$, with remarkable topological properties*.

It's certainly true that (B) $\to$ (A);
ideally then, we'd hope/expect that (A) $\to$ (B).
In reality, this convergence is frequently** obstructed by singular perturbations, boundary layers, and (crucially) turbulence.

> **This is the perspective typically taken by Arnold and Khesin, for example.*

## **A motivating analogy: *Allen–Cahn***

To understand the pathology and disctinction of this "inviscid limit", let's consider the Allen–Cahn equation, a simple phase separation model, in 1D:

$$
2\varepsilon u''  =  W'(u).
$$

This can be understood as a minimisation problem on the energy

$$
\int_\Omega \left[\varepsilon(u')^2 + W(u)\right].
$$

The functional $W(u)$ is generally nonlinear, typically a double-well potential $W(u) = (1-u^2)^2$.

To fix ideas, let's impose boundary conditions (BCs) $u(-1) = -1$ and $u(1) = 1$.
Up to a *very* small error ($\mathcal{O}[\text{exp}(- \tfrac{2}{\sqrt{\varepsilon}})]$) solutions look like

$$
u_\varepsilon \sim \text{tanh}\left(\frac{x}{\sqrt{\varepsilon}}\right).
$$

### **Case A: *$\varepsilon \to 0_+$***

As $\varepsilon \to 0_+$, this $\text{tanh}$ solution gets steeper and steeper;
the solution converges to a step function:

$$
u_\varepsilon \to \text{sgn}(x).
$$

A lot of "physics" is contained within that interface:
there's a quantifiable and meaningful concentration of energy there.

### **Case B: *$\varepsilon = 0$***

Why take this limit?
Let's just set $\varepsilon = 0$ in the Allen–Cahn equation immediately:

$$
0 = W'(u) = 4(1 - u^2)u  \qquad
\implies  \qquad
u \in \{-1, 0, 1\}.
$$

That doesn't make sense?
This is completely ill-posed!

### **The analogy**

The parameter $\varepsilon$ in Allen–Cahn is analogous to the viscosity $\nu = \tfrac{1}{\text{Re}}$ in Navier–Stokes.
In either case, we have a small parameter acting on the largest derivative: a "singularly perturbed problem".

A. **The inviscid limit of Navier–Stokes:**
This corresponds then to Allen–Cahn as $\varepsilon \to 0_+$.
Analogous to the steepening $\tanh$ profile, as $\nu \to 0_+$, velocity gradients (i.e.~vorticities) concentrate in arbitrarily thin boundary layers and thin filaments: vortices.
The "roughness" doesn't vanish;
it just becomes infinitely fine.

B. **The Euler equations:**
This correspond to taking $\varepsilon = 0$ in Allen–Cahn.
This *"ideal version of Allen–Cahn"* ($0 = W'(u) = 4(1 - u^2)u$) has solutions with nice properties that are easy to analyse ($u \in \{-1, 0, 1\}$) but the problem doesn't really make sense.
What if I have a BC that's not in $\{-1, 0, 1\}$?
Can solutions just jump randomly between those values?
By dropping the highest-order derivative completely, the type of the equation changes.
This is the same for Euler.
Solutions do exist to Euler and they behave in very interesting ways, just like they do for *"ideal Allen–Cahn"*.
However, the typical setting for Navier–Stokes with BCs is now meaningless.

## **Vortex reconnection**

A nice consequence of this is in the reconnection of vortex filaments, as we discussed in the group meeting.

### **Case A: *The inviscid limit of Navier–Stokes ($\nu \to 0_+$)***

Pretty much anything is possible in Navier–Stokes.
If we're wondering whether reconnection can occur, we need only remind ourselves that solutions must decay to zero as the energy $\tfrac{1}{2} \int_\Omega \|\mathbf{u}\|^2$ decays to zero.

As two vortex filaments approach, we get steep velocity gradients.
Just as the Allen-Cahn interface width shrinks like $\sqrt{\epsilon}$, the length scale of velocity gradients we see in Navier–Stokes shrinks like $\sqrt{\nu}$.
But even if this is small, it's non-zero.
At this length, the sharp gradients case the viscous term $\nu\Delta\mathbf{u}$ to balance the convective term $\mathbf{u}\cdot\nabla\mathbf{u}$, and reconnection is possible.

This viscous reconnection allows for a change in topology.
This happens pretty slowly in general, but it surely does happen.

### **Case B: *The Euler equations ($\nu = 0$)***

As a Lie advection equation on the vorticity, solutions to the Euler equations *cannot change the topology of the vortex field*.
In a solution to the Euler equation *(if it exists...)* reconnection is topologically impossible.

## **Weak solutions \& the Onsager conjecture**

All this *"Euler's ill-posed"* and *"reconnection is possible in the inviscid limit"* nonsense becomes can be made much more rigorous, with a careful consideration of the analysis of the Euler equations.

### **The Onsager conjecture and dissipative anomalies**

While a naïve reading of the Euler equations might make us think these equations should conserve energy, that's quite hard to prove for weak solutions.
In fact, Lars Onsager conjectured in 1949 that, for weak solutions to the Euler equations to conserve energy, they must possess a certain regularity, specically a Hölder exponent $\alpha > 1/3$.

Initial data that is *"rougher"* than $\alpha = 1/3$ is referred to as *"wild"* in the literature, which I think is a fun term we should use more!
You can fix ideas when you hear people refer to wild initial data by thinking of the step function in the ideal Allen–Cahn equation;
we don't know what's going on in that wild and rough jump, but we know it's going to play all sorts of tricks on our solutions.

Sure enough, weak Euler solutions under wild initial data can dissipate kinetic energy, despite having zero viscosity.
There is some kind of *"invisible dissipation"* that occurs within these sharp gradients.
This relates to the idea that the limit of Navier–Stokes is, in some sense, a *"wild, dissipative weak solution to Euler"*, not a conservative classical one.

### **Convex integration and non-uniqueness**

Just like the ideal Allen–Cahn equation proposed above having infinitely many solutions (jumping around $u \in \{-1, 0, 1\}$), we can construct a continuum of weak solutions to the Euler equations under wild initial data.
These constructions are less straightword forward than the Allen–Cahn constructions, using techniques I really do not understand called convex integration and Nash–Kuiper embeddings.

In fact, one can construct wild solutions that are initially at rest, spontaneously start moving, and then stop, completely destroyed energy conservation, and giving whatever energy profile you might want.
There's a physical implication here that, hidden in the roughness of your initial data, there's all sorts of behaviour, the manifestation of which you have complete agency over.

Understandably, these continua of solutions have been referred to in the literature as *"very weak"*.
The *"ghost of viscosity"* (a limit of $\nu \to 0_+$) is required to pick out the physically relevant solution.
