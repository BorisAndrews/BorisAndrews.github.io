Note: The Singular Limit—Euler vs. The Vanishing Viscosity of Navier-Stokes

Target Audience: PhD Students in Analysis/PDEs

Subject: Fluid Dynamics, Singular Perturbations, and Weak Solutions

1. Introduction: Two Different Objects

A common misconception in early fluid dynamics education is treating the Euler equations simply as the Navier-Stokes equations with $\nu = 0$. While formally true at the level of the PDE symbols, analytically, they represent two distinct regimes.

We must distinguish between:

The Euler Equations: An ideal system describing the geodesic flow on the group of volume-preserving diffeomorphisms (Arnold’s geometric picture).

The Inviscid Limit: The accumulation point(s) of the sequence of solutions $\{u^\nu\}$ to the Navier-Stokes equations as $\nu \to 0$.

Ideally, we want $(2) \to (1)$. In reality, this convergence is obstructed by singular perturbations, boundary layers, and turbulence.

2. A Motivating Analogy: Allen-Cahn

To understand the pathology of the inviscid limit, let us look at the Allen-Cahn equation, a model for phase separation. Consider the energy functional on the interval $[-1, 1]$:

$$E_\epsilon[u] = \int_{-1}^{1} \left( \frac{\epsilon}{2} |u'|^2 + \frac{1}{\epsilon} W(u) \right) dx$$

where $W(u) = (1-u^2)^2$ is a double-well potential. We impose Boundary Conditions (BCs): $u(-1) = -1$ and $u(1) = 1$.

Case A: The Limit of $\epsilon \to 0$

For $\epsilon > 0$, the competition between the gradient penalty (diffusion) and the potential forces the solution to resemble a $\tanh$ profile. It stays close to $-1$ on the left, close to $+1$ on the right, and transitions smoothly over a region of width $O(\epsilon)$.

As $\epsilon \to 0$, this transition layer (interface) steepens. The gradient $\nabla u$ blows up ($ \sim 1/\epsilon$), but the solution converges sensibly to a step function (the "sharp interface limit"). The "physics" here is contained in the interface; the energy concentrates there.

Case B: Setting $\epsilon = 0$ Directly

If we set $\epsilon = 0$ in the functional immediately, we lose the derivative term entirely. We are strictly minimizing $\int W(u)$. The solution is simply $u(x) \in \{-1, 1\}$ almost everywhere.
This is ill-posed:

It satisfies the equation $W'(u)=0$ everywhere, but implies nothing about where the jump occurs.

It cannot satisfy the BCs in a continuous sense (singular perturbation).

The Analogy:

Navier-Stokes ($\nu > 0$): Like Allen-Cahn with $\epsilon > 0$. The viscosity $\nu \Delta u$ regularizes the flow, enforcing no-slip boundaries via a boundary layer (analogous to the $\tanh$ transition) and smoothing out small scales.

The Inviscid Limit: Like the steepening $\tanh$. As $Re \to \infty$, gradients of velocity (vorticity) concentrate in boundary layers and thin filaments. The "roughness" does not vanish; it becomes infinitely fine.

The Euler Equations: Like the $\epsilon=0$ algebraic minimization. By dropping the highest-order derivative ($\Delta u$), we change the type of the equation. We lose the ability to enforce no-slip conditions, leading to non-uniqueness and ill-posedness.

3. Vortex Reconnection: Geometry vs. Analysis

This distinction resolves the apparent paradox of vortex reconnection.

The Euler View (Geometric)

If we view Euler strictly as the transport of vorticity:


$$\partial_t \omega + (u \cdot \nabla)\omega = (\omega \cdot \nabla)u$$


This is a Lie advection equation. By Kelvin’s Circulation Theorem and Helmholtz's laws, vortex lines are material lines. They move with the flow and cannot cut through each other. In a smooth classical solution to Euler, reconnection is topologically impossible.

The Navier-Stokes Limit View (Analytic)

In the physical world (Navier-Stokes with tiny $\nu$), reconnection does happen. Why?
As two vortex tubes approach, the strain field creates steep gradients in velocity. Just like the Allen-Cahn interface width shrinks with $\epsilon$, the length scale of the interaction shrinks with $\nu$.
Locally, the gradient $\nabla u$ becomes so large that the viscous term $\nu \Delta u$ becomes $O(1)$ even if $\nu \approx 0$.


$$\lim_{\nu \to 0} (\nu \Delta u) \neq 0 \quad \text{on the singular set.}$$


This viscous reconnection allows for a change in topology and energy dissipation. This suggests that the "physical" Euler solution (the limit of NS) should perhaps allow for reconnection, even if the "classical" PDE does not.

4. Shocks, Weak Solutions, and "Wild" Euler

You mentioned shock solutions. This is where the analogy to "Euler doesn't really work" becomes most rigorous.

In compressible flow, characteristics cross, leading to shocks (discontinuities in $u$). Standard derivatives fail, necessitating weak solutions. For the incompressible Euler equations, we don't have compressible shocks, but we have an analogous concept of "weak" turbulence.

The Onsager Conjecture and Dissipative Anomaly

Lars Onsager (1949) conjectured that for Euler solutions to conserve energy, they must have a certain regularity (Hölder exponent $\alpha > 1/3$).
If the flow is "rougher" than $\alpha = 1/3$ (analogous to the steep gradients in the Allen-Cahn limit), Euler solutions can dissipate kinetic energy despite having zero viscosity.
This confirms the idea that the limit of Navier-Stokes is a dissipative weak solution to Euler, not a conservative classical one.

Convex Integration and Non-Uniqueness

Recent breakthroughs by De Lellis and Székelyhidi (using techniques akin to Nash-Kuiper embedding) have shown that weak solutions to the Euler equations are spectacularly non-unique.
If we consider Euler without the selection principle of viscosity (like setting $\epsilon=0$ in Allen-Cahn):

There are infinitely many weak solutions for given initial data.

We can construct solutions that are initially at rest, spontaneously start moving, and then stop (breaking energy conservation).

This reinforces the view that Euler alone is under-determined. Just as minimizing $\int W(u)$ in Allen-Cahn doesn't tell you where the interface is, the Euler equations alone don't tell you how to handle the "infinite frequency" turbulence. We need the "ghost of viscosity" (the limit of $\nu \to 0$) to select the physically relevant solution.
